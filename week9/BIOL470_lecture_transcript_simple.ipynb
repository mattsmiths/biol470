```python
phrase = "cool"




nFinal = []

# going through and opening all lectures
for ind,lecture1 in enumerate(lecture_dates):
  in1 = open(lecture1,'r')
  out = in1.readlines()
  in1.close()

  # for each line in the transcript
  nCount = 0
  for line in out:

    # identify if the phrase is present
    if phrase.lower() in line.lower():
      nCount += 1

  # add the final count to a list
  nFinal.append(nCount)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(np.arange(len(nFinal)),nFinal)

plt.ylabel('Word count')
plt.xlabel('Date')

tempname = [ele.split('/')[-1].split('.')[0] for ele in lecture_dates]
plt.xticks(np.arange(len(nFinal)),tempname,rotation=80)
plt.show()
```
